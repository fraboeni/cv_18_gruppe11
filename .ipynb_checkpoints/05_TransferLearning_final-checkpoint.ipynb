# Übung 6: Transfer learning

Neuronal Netze benötigen oft eine große Menge an Trainingsdaten, damit es nicht zu overfitting kommt. Transfer learning erlaubt es, mit relativ geringen Datenmenge dennoch erfolgreiche große Netze zu trainieren. Dabei verwendet man ein bereits auf einen anderen Datensatz (z.b. ImageNet) vortrainiertes Netzwerk, und ersetzt nur das letzte Layer durch ein neues. In dieser Übung geht es darum, ein Netzwerk für die Erkennung von Geparden und Leoparden in der freien Wildbahn zu trainineren. 

## Daten laden

Lade die Daten hier herunter: http://tonic.imp.fu-berlin.de/cv_data/data.tar.gz

Die Daten wurde bereits in ein Trainings- und Validierungsset geteilt. Die Ordnerstruktur ist wie bei vielen Bildklassifierungsdatensetzen so aufgebaut. Es gibt zwei Unterordner für die Trainings- und Validierunsdaten. In diesen Ordnern liegen dann jeweils alle Bilder von einer Klasse in einem Unterordner mit dem Namen der Klasse.

Ein Beispiel: Die Trainingsbilder für die Klasse "cheetah" liegen in dem Unterordner train/cheetah

Diese Orderstruktur wird auch von dem in keras enhaltenen ImageDataGenerator unterstützt.

```python
import tensorflow as tf
from keras.backend.tensorflow_backend import set_session, get_session
config = tf.ConfigProto()
config.gpu_options.allow_growth = True
set_session(tf.Session(config=config))

import os
from keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```python
batch_size = 32
image_input_size = (224, 224)
# small_image_input_size = (32, 32)
data_path = r'C:\Users\Eduard Gette\Documents\Computer Vision\data'
```

```python
train_data_path = os.path.join(data_path, 'train')
val_data_path = os.path.join(data_path, 'val')

izw_classes = ('unknown', 'cheetah', 'leopard')

generator = ImageDataGenerator(horizontal_flip=True, rescale=1/255.0)
val_generator = ImageDataGenerator(horizontal_flip=False, rescale=1/255.0)

train_gen = generator.flow_from_directory(
    train_data_path, 
    target_size=image_input_size,
    classes=izw_classes,
    batch_size=batch_size)

val_gen = val_generator.flow_from_directory(
    val_data_path, 
    target_size=image_input_size,
    classes=izw_classes,
    batch_size=batch_size)
```

## Training ohne transfer learning

Trainiere zuerst ein kleines Classifer-Netzwerk ohne transfer learning. Falls du keine Grafikkarte hast, solltest du nicht die volle Auflösung (siehe Variable image_input_size) verwenden, da das Training sonst zu lange dauert. Eine Bildgröße von 32x32 Pixeln wäre zum Beispiel möglich.

Structure of train_gen:
    - bunch of tuples, one for each batch according to batchsize
        - tuple 1st position: datatensor
        - tuple 2nd position: labels one-hot encoded

```python
from skimage import io
# see whether image still look reasonable
example_img = train_gen[1][0][0,:,:]
io.imshow(example_img)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Input, GlobalAveragePooling2D
from keras.layers import Activation, Dropout, Flatten, Dense

model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(image_input_size[0], image_input_size[1], 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(3))
model.add(Activation('sigmoid'))

model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
generator = ImageDataGenerator(rescale=1/255.0)
pred_gen = generator.flow_from_directory(
        val_data_path,
        target_size=(224, 224),
        classes=izw_classes,
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)
predictions = model.predict_generator(pred_gen)
```

```python
def get_onehot(predictions):
    predictions = (predictions==predictions.max(axis=1)[:,None]).astype(int)
    return predictions

one_hot_preds = get_onehot(predictions)
one_hot_preds
```

```python
import numpy as np
y_pred = np.argmax(predictions, axis=1)
generator2 = ImageDataGenerator(rescale=1/255.0)
true_y_gen = generator2.flow_from_directory(
        val_data_path,
        target_size=(224, 224),
        classes=izw_classes,
        batch_size=batch_size,
        shuffle=False)

y_true = []
for i in range(len(true_y_gen)):
    b_tuple = true_y_gen[i]
    labels = b_tuple[1]
    int_labels = list(np.argmax(labels, axis=1))
    y_true += int_labels
y_true = np.array(y_true)
y_true
```

Ersetze dafür das letzte Layer in dem Netzwerk mit einem Dense Layer mit 3 Ausgaben für unsere Klassen cheetah, leopard und unknown. Du kannst selbst entscheiden, ob du nun das komplette Netzwerk mit trainierst oder nur das neu eingefügte, letzte Layer.

```python
train_labels = np.load('train_labels.npy')
print(train_labels.shape)
train_labels
```

```python
# Quelle: https://stackoverflow.com/questions/4601373/better-way-to-shuffle-two-numpy-arrays-in-unison?utm_medium=organic&utm_source=google_rich_qa&utm_campaign=google_rich_qa
def shuffle_in_unison(a, b):
    assert len(a) == len(b)
    shuffled_a = np.empty(a.shape, dtype=a.dtype)
    shuffled_b = np.empty(b.shape, dtype=b.dtype)
    permutation = np.random.permutation(len(a))
    for old_index, new_index in enumerate(permutation):
        shuffled_a[new_index] = a[old_index]
        shuffled_b[new_index] = b[old_index]
    return shuffled_a, shuffled_b
```

```python
# shuffle data for training top network
X_shuffled, y_shuffled = shuffle_in_unison(bottleneck_features, train_labels)
```

```python
# X_shuffled = np.expand_dims(X_shuffled, axis=0)
```

```python
input_shape = X_shuffled.shape
input_shape
```

```python
# build a classifier model to put on top of the convolutional model
top_model = Sequential()
top_model.add(Dense(1024, activation='relu', input_shape=input_shape[1:]))
top_model.add(Dropout(0.5))
top_model.add(Dense(512, activation='relu'))
top_model.add(Dropout(0.5))
top_model.add(Dense(256, activation='relu'))
top_model.add(Dropout(0.5))
top_model.add(Dense(3, activation='sigmoid'))

top_model.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
```

```python
top_model.fit(x=X_shuffled, y=y_shuffled, verbose=1, validation_split=0.2,
              batch_size=batch_size, epochs=20, shuffle=True, steps_per_epoch=None)
```

- After 20 epochs already 98% accuracy on the validation split of the training data
- Now we save the weights and validate on the actual validation data just to make sure everything is correct

```python
# create necessary generators
generator = ImageDataGenerator()
val_gen = generator.flow_from_directory(
    val_data_path, 
    target_size=image_input_size,
    classes=izw_classes,
    batch_size=batch_size,
    shuffle=False)
```

```python
# bottleneck_features_val = model.predict_generator(val_gen)
```

```python
bottleneck_features_val = np.load("bottleneck_features_val")
```

```python
# get the correct labels for the validation data set
val_labels = np.ndarray(shape=(bottleneck_features_val.shape[0],3))
ind = 0
for i in range(len(val_gen)):
    y = val_gen[i][1]
    val_labels[ind:ind+batch_size,:] = y
    ind += batch_size
```

```python
X_val_shuffled, y_val_shuffled = shuffle_in_unison(bottleneck_features_val, val_labels)
```

```python
# make predictions for the val_data
predictions = top_model.predict(X_val_shuffled, batch_size=batch_size, verbose=1)
```

```python
predictions_scalar = np.argmax(predictions, axis=1)
predictions_scalar
```

```python
val_labels_scalar = np.argmax(y_val_shuffled, axis=1)
val_labels_scalar
```

Evaluiere das so trainierte Netzwerk wie in den letzten beiden Aufgaben.

```python
true_labels = pd.Series(val_labels_scalar)
pred_labels = pd.Series(predictions_scalar)

pd.crosstab(true_labels, predictions_scalar, rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
# like before roc auc score only for cheetah therefore it might be that it doesn't seem to be reasonably
# better than the simple non pretrained model, but most of the improvement of the pretrained model
# comes from correctly classifying leopards (see misclassifications of leopards as gepards in the according 
# confusion matrix - 72 vs. only 32)
y_pred_cheetah = predictions[:, 1]
y_true_cheetah = y_val_shuffled[:,1]
roc_auc_score(y_true_cheetah, y_pred_cheetah)
```
